##TARGET VS NUMERIC (AGE)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 8))

for label in data['Label'].unique():
    subset = data[data['Label'] == label]
    sns.histplot(subset['Age'], kde=True, label=label, bins=10, alpha=0.5)

plt.title('Distribution of Age by Labels with KDE', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend(title='Label')
plt.show()
```

##TARGET VS CATEGORICAL/ORDINAL

```python
cate_cols=['Age','Gender','Weight','Height','BMI']

fig, axs=plt.subplots(2,3,figsize=(16,6))

for col,ax in zip(cate_cols,axs.flatten()):
  sns.countplot(data=data,x=col,hue='Label', ax=ax)
```
